```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/Users/hangvu/Documents/school/nlp/project/INT3406-GROUP5-minBERT_for_document_classification/minbert-75-0.001/run.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Plot losses
plt.figure(figsize=(8, 6))
plt.plot(data['epoch'], data['train_loss'], label='Training Loss', marker='o')
plt.plot(data['epoch'], data['dev_loss'], label='Validation Loss', marker='o')
plt.title('Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plot accuracies
plt.figure(figsize=(8, 6))
plt.plot(data['epoch'], data['train_acc'], label='Training Accuracy', marker='o')
plt.plot(data['epoch'], data['dev_acc'], label='Validation Accuracy', marker='o')
plt.title('Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
