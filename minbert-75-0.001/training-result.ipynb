```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/Users/hangvu/Documents/school/nlp/project/INT3406-GROUP5-minBERT_for_document_classification/minbert-75-0.001/run.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Interpolated plot for loss
epochs = data['epoch']
train_loss = data['train_loss']
val_loss = data['dev_loss']

plt.figure(figsize=(8, 6))
plt.plot(epochs, train_loss, label='Training Loss', linestyle='-')
plt.plot(epochs, val_loss, label='Validation Loss', linestyle='-')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
